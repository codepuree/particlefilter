## Setup matlab environment

```matlab
clear all;
close all;
clc;

cd('../src');
```

## Load map

```matlab
map = LoadMap('../Data/Vorgabe_update.png');

show(map);
```

## Generate raster

```matlab
[X, Y] = meshgrid(map.XWorldLimits(1):0.5:map.XWorldLimits(2), ...
    map.YWorldLimits(1):0.5:map.YWorldLimits(2)...
);
X      = reshape(X,numel(X),1);
Y      = reshape(Y,numel(Y),1);

mesh   = [X,Y];
```

## Get unoccupied coordinates from raster

```matlab
rasterOccupancy = getOccupancy(map, mesh);

rIdxOcc   = rasterOccupancy < 0.5;
rIdxUnOcc = rasterOccupancy > 0.5;
coorsFree = mesh(rIdxOcc, :);
% die       = mesh(rIdxUnOcc, :);
```

## Visualize

```matlab
show(map);
hold on;
scatter(coorsFree(:, 1), coorsFree(:, 2), '*r');
```
