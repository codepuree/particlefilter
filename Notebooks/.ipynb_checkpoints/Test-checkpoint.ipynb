# Load the pointCloud

```matlab
ptCloud = pcread('../Data/ptCloudOriginal.ply');
```

# Display point cloud

```matlab
pcshow(ptCloud, 'VerticalAxis', 'y', 'VerticalAxisDir', 'down');
daspect([1, 1, 1]);
```

# Filter cross section

Filter between -0.2 m to 0.2 m

```matlab
ptCloudNew = GetCrossProfile(ptCloud, -0.2, 0.2)
```

```matlab
pcshow(ptCloudNew, 'VerticalAxis', 'Y') % , 'VerticalAxisDir', 'down')
daspect([1, 1, 1]);
title('Cross section');
xlabel('X');
ylabel('Y');
zlabel('Z');
```

# Flatten cross section

```matlab
ptCloudFlattend = FlattenPointCloud(ptCloudNew)
```

```matlab
pcshow(ptCloudFlattend, 'VerticalAxis', 'Y');
daspect([1, 1, 1]);
title('Flattend profile')
xlabel('X')
ylabel('Y')
zlabel('Z')
```

### Alternative visualization

```matlab
scatter(ptCloudFlattend.Location(:, 1), ptCloudFlattend.Location(:, 3), '.');
daspect([1, 1, 1]);
title('Quer-Profile')
xlabel('X')
ylabel('Y')
```

## Save Data

```matlab
pcwrite(ptCloudFlattend, '../Data/ptCloudFlattend.ply');
```

```matlab
class(ptCloudFlattend.Location)
```

```matlab
map = [ ptCloudFlattend.Location(:, 1), ptCloudFlattend.Location(:, 3) ];
size(map)
```

### Get Grid dimensions

```matlab
xMin = min(map(:, 1));
xMax = max(map(:, 1));
yMin = min(map(:, 2));
yMax = max(map(:, 2));

width  = abs(abs(xMax) - abs(xMin));
height = abs(abs(yMax) - abs(yMin));
```

```matlab
disp(['xMin:   ' num2str(xMin)]);
disp(['xMax:   ' num2str(xMax)]);
disp(['width:  ' num2str(width) 10]);

disp(['yMin:   ' num2str(yMin)]);
disp(['yMax:   ' num2str(yMax)]);
disp(['height: ' num2str(height)]);
```

### Convert to polar coordinates

```matlab
[thetas, rhos] = cart2pol(ptCloudFlattend.Location(:, 1), ptCloudFlattend.Location(:, 3));
```

```matlab
polarplot(thetas, rhos, '.');
title('Polar coordinates');
```

#### Polar coordinates statistics

```matlab
thetaMin  = min(thetas);
thetaMax  = max(thetas);

rhoMin    = min(rhos);
rhoMax    = max(rhos);
rhoMean   = mean(rhos);
rhoMedian = median(rhos);

thetaDiff = thetaMax - thetaMin;
rhoDiff   = rhoMax - rhoMin;
```

```matlab
disp(['thetaMin:  ' num2str(thetaMin)  ' rad (' num2str(rad2deg(thetaMin))  ' deg)']);
disp(['thetaMax:  ' num2str(thetaMax)  ' rad (' num2str(rad2deg(thetaMax))  ' deg)']);
disp(['thetaDiff: ' num2str(thetaDiff) ' rad (' num2str(rad2deg(thetaDiff)) ' deg)' 10])

disp(['rhoMin:    ' num2str(rhoMin)    ' m']);
disp(['rhoMax:    ' num2str(rhoMax)    ' m']);
disp(['rhoDiff:   ' num2str(rhoDiff)   ' m']);
disp(['rhoMean:   ' num2str(rhoMean)   ' m']);
disp(['rhoMedian: ' num2str(rhoMedian) ' m']);
```

### Reduce polar coordinates

```matlab
thetaStep = thetaDiff / 500
```

```matlab
[thetasSorted, thetasSortedIndex] = sort(thetas);
rhosSorted = rhos(thetasSortedIndex);
```

```matlab
[thetasUnique, thetasUniqueIndex, ~] = unique(thetas);
rhosUnique = rhos(thetasUniqueIndex);
```

```matlab
disp(['      thetas: ' num2str(numel(thetas))]);
disp(['thetasUnique: ' num2str(numel(thetasUnique))]);
```

```matlab
polarplot(thetasUnique, rhosUnique, '.');
title('Polar coordinates (unique thetas)');
```

```matlab
thetasUnique  = unique(thetas);

rhosUniqueMin    = arrayfun(@(x)    min(sort(rhos(thetas == x), 'descend')), thetasUnique);
rhosUniqueMax    = arrayfun(@(x)    max(sort(rhos(thetas == x), 'descend')), thetasUnique);
rhosUniqueMean   = arrayfun(@(x)   mean(sort(rhos(thetas == x), 'descend')), thetasUnique);
rhosUniqueMedian = arrayfun(@(x) median(sort(rhos(thetas == x), 'descend')), thetasUnique);
```

```matlab
% Custom resample to 500 elements
thetasSampled = linspace(thetaMin, thetaMax, 1000);
rhosSampled   = arrayfun(@(x) median(rhos(thetas >= x & thetas <= (x + thetaDiff))), thetasSampled);
```

```matlab
figure('name', 'simple');

ax1 = subplot(1, 2, 1);
scatter(ptCloudFlattend.Location(:, 1), ptCloudFlattend.Location(:, 3), '.');
pbaspect(ax1,[1 1 1]);
title(['Quer-Profile' 10 'points: ' num2str(numel(ptCloudFlattend.Location(:, 1)))])
xlabel('X')
ylabel('Y')

ax2 = subplot(1, 2, 2);
[x, y] = pol2cart(thetasUnique, rhosUnique);
scatter(x, y, '.');
pbaspect(ax2,[1 1 1]);
title(['unique thetas' 10 'points: ' num2str(numel(rhosUnique))])
xlabel('X')
ylabel('Y')

figure('name', 'min/max');

ax1 = subplot(1, 2, 1);
[x, y] = pol2cart(thetasUnique, rhosUniqueMin);
scatter(x, y, '.');
pbaspect(ax1,[1 1 1]);
title(['Min rhos' 10 'points: ' num2str(numel(rhosUniqueMin))])
xlabel('X')
ylabel('Y')

ax2 = subplot(1, 2, 2);
[x, y] = pol2cart(thetasUnique, rhosUniqueMax);
scatter(x, y, '.');
pbaspect(ax2,[1 1 1]);
title(['Max rhos' 10 'points: ' num2str(numel(rhosUniqueMax))])
xlabel('X')
ylabel('Y')

figure('name', 'mean/median');

ax1 = subplot(1, 2, 1);
[x, y] = pol2cart(thetasUnique, rhosUniqueMean);
scatter(x, y, '.');
pbaspect(ax1,[1 1 1]);
title(['Mean rhos' 10 'points: ' num2str(numel(rhosUniqueMean))])
xlabel('X')
ylabel('Y')

ax2 = subplot(1, 2, 2);
[x, y] = pol2cart(thetasUnique, rhosUniqueMedian);
scatter(x, y, '.');
pbaspect(ax2,[1 1 1]);
title(['Median rhos' 10 'points: ' num2str(numel(rhosUniqueMedian))])
xlabel('X')
ylabel('Y')

figure('name', 'sample');

ax1 = subplot(1, 2, 1);
scatter(ptCloudFlattend.Location(:, 1), ptCloudFlattend.Location(:, 3), '.');
pbaspect(ax1,[1 1 1]);
title(['Quer-Profile' 10 'points: ' num2str(numel(ptCloudFlattend.Location(:, 1)))])
xlabel('X')
ylabel('Y')

ax2 = subplot(1, 2, 2);
[x, y] = pol2cart(thetasSampled, rhosSampled);
scatter(x, y, '.');
pbaspect(ax2,[1 1 1]);
title(['unique thetas' 10 'points: ' num2str(numel(rhosSampled))])
xlabel('X')
ylabel('Y')
ylim([0, 4])
```
